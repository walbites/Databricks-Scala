```
%scala
import org.apache.spark.sql.types._

val path = "/FileStore/INICIAL/ubigeo.csv"

val schema = new StructType()
  .add("UBIGEO",StringType,true)
  .add("DISTRITO",StringType,true)
  .add("PROVINCIA",StringType,true)
  .add("DEPARTAMENTO",StringType,true)
  .add("LATITUD",FloatType,true)
  .add("LONGITUD",FloatType,true)

val df_ubigeo = spark.read.format("csv")
  .option("header", "true")
  .schema(schema)
  .load(path)
```
