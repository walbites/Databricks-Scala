###### 1. Creación de una Clase Persona MUTABLE, con dni, nombre, telefono y edad. Los campos van definidos en el bloque, utilizando los argumentos del constructor.

```
%scala
class Persona(_dni:String,_nombre:String,_telefono: String, _edad: Int){
  var dni: String= _dni
  var nombre: String= _nombre
  var telefono: String = _telefono
  var edad: Int = _edad
}
```

```
%scala
println(persona1.dni)
println(persona1.nombre)
println(persona1.telefono)
println(persona1.edad)
```

```
%scala
println(persona1.edad)
```

######4. Podemos declarar constructores adicionales mediante def this:

```
%scala
class Persona(val dni:String,val nombre:String,val telefono: String,val edad: Int){
  def this(_dni:String, _nombre:String, _telefono:String)={
    this(_dni,_nombre,_telefono,0)
  }
}
```

```
%scala
println(persona4.dni)
println(persona4.nombre)
println(persona4.telefono)
println(persona4.edad)
```

######5. Los métodos se añaden utilizando def. Podemos, por ejemplo, permitir que nuestra persona pueda disminuir su edad  dado un decremento.

```
%scala
class Persona(val dni:String,val nombre:String,val telefono: String,val edad: Int){
  def dismunirEdad(decremento:Int): Persona={
    new Persona(dni,nombre,telefono,edad-decremento)
  }
}
```

```
%scala
val persona5= new Persona("8888888","Alberto","86767464",40)
```

```
%scala
println(persona5.dni)
println(persona5.nombre)
println(persona5.telefono)
println(persona5.edad)
```

```
%scala
val persona5D= persona5.dismunirEdad(5)
```

```
%scala
println(persona5D.dni)
println(persona5D.nombre)
println(persona5D.telefono)
println(persona5D.edad)
```

######6. Es posible tener clases con una única instancia, lo que se conoce como Singleton Objects. Este componente nos permite reemplazar los miembros de clase propios de Java. Crearemos un Objeto Persona con valores iniciales y un método de creación de persona crear.

```
%scala
object CrudPersona{
  def crearPersona(dni:String,nombre:String,telefono: String,edad: Int): Persona={
    new Persona(dni,nombre,telefono,edad)
  }
}
```

```
%scala
val cp= CrudPersona.crearPersona("857575","Maria","774464",25)
```

```
%scala
println(cp.dni)
println(cp.nombre)
println(cp.telefono)
println(cp.edad)
```

######7. Podemos aplicar herencia basica entre clases, definamos una clase Persona Juridica que hereda de nuestra clase Persona añadiendo un campo ruc

```
%scala
class PersonaJuridica(dni:String,nombre:String,telefono: String,edad: Int, val ruc: String)
extends Persona(dni,nombre,telefono,edad)
```

```
%scala
val pj= new PersonaJuridica("03373734734","Carmen","9867574",50,"1103373734734")
```

```
%scala
println(pj.dni)
println(pj.nombre)
println(pj.telefono)
println(pj.edad)
println(pj.ruc)
```

######8. Los Traits son interfaces que permiten la implementacion parcial, con lo que se permite la herencia multiple

```
%scala
trait PersonaExtranjera{
  val pasaporte: String="7777777777"
  val pais: String="Venezuela"
}
```

```
%scala
class MaestraPersona(dni:String,nombre:String,telefono: String,edad: Int) 
extends Persona (dni,nombre,telefono,edad)
with PersonaExtranjera
```

```
%scala
val mp= new MaestraPersona("-","nttdata","9999",10)
```

```
%scala
println(mp.dni)
println(mp.nombre)
println(mp.telefono)
println(mp.edad)
println(mp.pasaporte)
println(mp.pais)
```

######9. Una clase Abstract. solo necesita usar cuando: Quieres crear una clase base que requiera argumentos de constructor o se requiera invocar a su código Scala desde el código Java

```
%scala
abstract class NombresCompletos(nombres:String, paterno:String, materno:String){
  def mostrarNombresCompletos()
}
```

```
%scala
class Nombres(nombres:String, paterno:String, materno:String) extends NombresCompletos(nombres,paterno,materno){
  def mostrarNombresCompletos(){
    println("Nombres: "+ nombres)
    println("Paterno: "+ paterno)
    println("Materno: "+ materno)
  }
}
```

```
%scala
val nombres = new Nombres("Walter","Albites","Azarte")
```

```
%scala
nombres.mostrarNombresCompletos()
```
